```python
import matplotlib.pyplot as plt
import numpy as np

res = np.array([[64644027, 65173009, 67021737],
       [74693813, 73329447, 75679921],
       [54907087, 56132980, 54979398],
       [34923365, 26214941, 33430824],
       [11864729, 14258778, 12614188],
       [5183144, 5334029, 4936388]])

thread_num = np.array([1, 2, 4, 8, 16, 32])
```

```python
res[:, 0]
```

```python
default_thread_num_ticks = range(len(thread_num))

plt.plot(default_thread_num_ticks, res[:, 0])
plt.xlabel('number of threads')
plt.ylabel('update count')
plt.xticks(default_thread_num_ticks, thread_num)
plt.show()
```
